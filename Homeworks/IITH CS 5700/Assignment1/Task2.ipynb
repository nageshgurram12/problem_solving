<h3> Plot the graph for terms and tokens in huge corpus of reviews : <h3> <hr/>

```python
import gzip

data_file =  "./data/reviews_Tools_and_Home_Improvement_5.json.gz" # consists of 134K reviews
```

```python
def parse(path):
    '''
    Parse the file in path and give line by line
    '''
    g = gzip.open(path, 'r')
    for l in g:
        review = eval(l)
        yield review
```

```python
from nltk.tokenize import word_tokenize
import hyperloglog

# Use hyperloglog counter to count distict tokens with 1% error
hll = hyperloglog.HyperLogLog(0.01)  
```

```python
def get_terms_and_tokens(file_path):
    '''
    Process each review and print tokens and terms after
    1K reviews are done, in the end plot them
    '''
    tokens = 0
    terms = 0
    tokens_terms = {}
    review_gen = parse(file_path)
    proc_reviews = 0 # Processed reviews till now
    for review in review_gen:
        proc_reviews += 1
        for token in word_tokenize(review["reviewText"]):
            token = token.lower()
            token.strip()
            if token == '':
                continue
            tokens += 1
            
            hll.add(token)
            
            if tokens % 10000 == 0:
                tokens_terms[tokens] = len(hll)
                print("Review no: " + str(proc_reviews) + " tokens : " + str(tokens) +
                      " terms: " + str(tokens_terms[tokens]))
        
    return tokens_terms
```

```python
tokens_terms = get_terms_and_tokens(data_file)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.log10(list(tokens_terms.keys())), np.log10(list(tokens_terms.values())), 
         label="Log(Tokens) Vs Log(Terms)")
plt.xlabel("Log(Tokens)")
plt.ylabel("Log(Terms)")
```
