```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xgb
```

```python
input_data = pd.read_csv("train.csv", index_col="ID")
input_data.head()
```

```python
test_samples = pd.read_csv("test.csv", index_col="ID")
test_samples.head()
```

```python
input_features = input_data.iloc[:,:-1].values
input_classes = input_data.iloc[:,-1].values

(train_samples, val_samples, train_classes, val_classes) = \
train_test_split(input_features, input_classes, test_size=0.2, stratify=input_classes, random_state=0)
```

```python
def predict_accuracy(model):
    print(model.best_params_, model.best_score_)
    predict_classes = model.predict(val_samples)
    print("Accuracy : %0.4f" % accuracy_score(val_classes, predict_classes))
```

```python
grid_parameters = {'n_estimators':range(130,150,10)}
grid_xgb = GridSearchCV(xgb.XGBClassifier(max_depth=4),  grid_parameters, cv=3, n_jobs=-1)
grid_xgb.fit(train_samples, train_classes)
```

```python
predict_accuracy(grid_xgb)
```

```python
grid_parameters = {"learning_rate": [0.2, 0.3, 0.4, 0.5]}
grid_xgb = GridSearchCV(xgb.XGBClassifier(max_depth=4, n_estimators=140),  grid_parameters, cv=3, n_jobs=-1)
grid_xgb.fit(train_samples, train_classes)
```

```python
predict_accuracy(grid_xgb)
```

```python
grid_parameters = {'min_child_weight':range(1,6,1), 'max_depth':range(1,4,1), 'gamma' : [i/10.0 for i in range(0,5)]}
grid_xgb = GridSearchCV(xgb.XGBClassifier(n_estimators=140, learning_rate=0.3),  grid_parameters, cv=3, n_jobs=-1)
grid_xgb.fit(train_samples, train_classes)
```

```python
predict_accuracy(grid_xgb)
```

```python
grid_parameters = {'gamma' : [i/10.0 for i in range(0,5)]}
grid_xgb = GridSearchCV(xgb.XGBClassifier(n_estimators=140, learning_rate=0.3, \
                                          min_child_weight=, max_depth=4), \
                        grid_parameters, cv=3, n_jobs=-1)
grid_xgb.fit(train_samples, train_classes)
```

```python
predict_accuracy(grid_xgb)
```

```python
xgb_tree = xgb.XGBClassifier(n_estimators=130, subsample=0.5)
xgb_tree.fit(train_samples, train_classes)
```

```python
predict_accuracy(xgb_tree)
```
