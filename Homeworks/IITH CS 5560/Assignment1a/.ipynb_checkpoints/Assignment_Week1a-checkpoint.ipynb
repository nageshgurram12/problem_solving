## Dataset: KEGG Metabolic Relation Network (Directed) Data Set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
from scipy.optimize import fsolve
import matplotlib.backends.backend_pdf
from scipy.stats import multivariate_normal
from scipy.stats import dirichlet
pdf = matplotlib.backends.backend_pdf.PdfPages("plots.pdf")
```

```python
dataset = pd.read_csv("testdata.csv", header=None)
dataset = dataset.iloc[:, 1:]
```

Devide the data into k parts:

```python
def divide_array_into_k_parts(data, k=10):    
    total_elements = data.size
    if(k > total_elements):
        return data
    partition_size = int(total_elements/k)
    if total_elements % k == 0:
        return np.reshape(data, (k, partition_size))
    else:
        return np.reshape(data[:k*partition_size], (k, partition_size))
```

Plot the parameters with the size of distribution:

Use two Y-axis to show for two paramters

```python
def plot_params(data_sizes, params, param_names, title):
    fig, ax1 = plt.subplots()
    ax1.plot(data_sizes, params[0], 'g', label=param_names[0])
    ax1.set_xlabel("Data size")
    ax1.set_ylabel(param_names[0])
    ax1.legend(loc="upper right")
    if len(param_names) > 1:
        ax2 = ax1.twinx()
        ax2.plot(data_sizes, params[1], 'b', label=param_names[1])
        ax2.set_ylabel(param_names[1])
        ax2.legend(bbox_to_anchor=(1,0.9))
    plt.title(title)
    pdf.savefig(fig)
    plt.show()
```

Remove the outliers by taking k std deviations away from mean

```python
def remove_outliers(data, k_std=3):
    d_mean = data.mean()
    d_std = data.std()
    return data[np.abs(data-d_mean) <= k_std*d_std]
```

Divide the datasets into k partitions and estiamte paramters for cumulative datasets

```python
def estimate_params(samples, param_estimator_func, params_names, title, num_partitions=100):
    dataset_sizes = []
    params_sets = [[] for i in range(len(params_names))]
    cum_data_set = np.array([]);
    data_sets = divide_array_into_k_parts(samples.values, num_partitions)
    for dset in data_sets:
        cum_data_set = np.append(cum_data_set, dset)
        params = param_estimator_func(cum_data_set)
        for i,x in enumerate(params):
            params_sets[i].append(x)
        dataset_sizes.append(cum_data_set.size)
    plot_params(dataset_sizes, params_sets, params_names, title)
```

### 1) Estimating parameters of Bernouli distribution:
 
 For 5th column in data set: Network Radius integer (min:1, max:2)  
 
 As the R.V takes values 1 and 2, I'm subtracting 1 from it to make it Bernouli

```python
bernouli_data = dataset[5]-1
def bernouli_param_estimator(dataset):
    return [dataset.mean()]
estimate_params(bernouli_data, bernouli_param_estimator, ["theta"], "Bernouli distribution")
```

### 2) Estimating paramters of Binomial distribution:

For column in dataset e) Network Diameter integer (min:1, max:30) 

```python
binomial_data_raw = dataset[4]
binomial_data = remove_outliers(binomial_data_raw, k_std=2)
def binomial_params_estimator(samples):
    bi_mean = samples.mean()
    bi_var = samples.var()
    theta = (1- bi_var/bi_mean)
    n = bi_mean/theta
    return [n, theta]
binomial_data.describe()
estimate_params(binomial_data, binomial_params_estimator, ["n", "theta"], "Binomial distribution")
```

### 3) Estimating paramters for Poisson distribution:

For column in dataset e) Network Diameter integer (min:1, max:30) 

```python
poisson_data_raw = dataset[4]
poisson_data = remove_outliers(poisson_data_raw)
def possion_param_estimator(samples):
    return [samples.mean()]
estimate_params(poisson_data, possion_param_estimator, ["delta"], "Poisson distribution")
```

### 4)  Estimating paramters for Normal distribution:

For column in data set - Avg.num.Neighbours real (min:1, [Web Link]) 

```python
normal_data_raw = dataset[8]
normal_data = remove_outliers(normal_data_raw)
def normal_param_estimator(samples):
    return [samples.mean(), samples.var()]
estimate_params(normal_data, normal_param_estimator, ["mean", "variance"] , "Normal distribution")
```

### 5) Estimating parameters student t distribution:

For column in data set - Avg.num.Neighbours real (min:1, [Web Link]) 

```python
student_t_data_raw = dataset[8]
student_t_data = remove_outliers(student_t_data_raw)
def student_t_param_estimator(samples):
    mean = samples.mean()
    df = samples.size - 1 # degree of freedom is number of samples - 1
    var = samples.var()
    scale = ((df-2) * var) / df
    return [mean, scale]
estimate_params(student_t_data, student_t_param_estimator, ["mean", "scale"], "student t distribution")
```

### 6) Estimating parameters for Laplace distribution:

For column in data set - Avg.num.Neighbours real (min:1, [Web Link]) 

```python
laplace_data_raw = dataset[8]
laplace_data = remove_outliers(laplace_data_raw)
def laplace_param_estimator(samples):
    mean = samples.mean()
    scale = math.sqrt(samples.var()/2)
    return [mean, scale]
estimate_params(laplace_data, laplace_param_estimator, ["mean", "scale"], "Laplace distribution")
```

### 7) Estimating paramters for Gamma distribution:

For column in data set - Avg.num.Neighbours real (min:1, [Web Link]) 

```python
gamma_raw = dataset[8]
gamma_data = remove_outliers(gamma_raw)
def gamma_param_estimator(samples):
    mean = samples.mean()
    var = samples.var()
    shape = (mean ** 2) / var
    rate = mean / var;
    return [shape, rate]
estimate_params(gamma_data, gamma_param_estimator, ["shape", "rate"], "Gamma distribution")
```

### 8) Estimating paramters for Inverse Gamma distribution:

For column in data set - Avg.num.Neighbours real (min:1, [Web Link]) 

```python
inverse_gamma_raw = dataset[8]
inverse_gamma_data = remove_outliers(inverse_gamma_raw)
def inv_gamma_equations(p, *args):
    (a,b) = p
    (m,v) = args
    return (b/(a-1) - m, b**2/((a-1)**2 * (a-2)) - v)

def inverse_gamma_param_estimator(samples):
    mean = samples.mean()
    var = samples.var()
    (a, b) = fsolve(inv_gamma_equations, (0.5,0.5), args=(mean, var), xtol=0.0001)
    return [a,b]
estimate_params(inverse_gamma_data, inverse_gamma_param_estimator, ["shape", "scale"], "Inverse Gamma distribution")
```

### 9) Estimating paramters for Exponetial distribution:

### 12) Estimating parameters of Beta distribution:

For column in dataset v) ClosenessCentrality real ([Web Link], max:1) 

```python
beta_raw_data = dataset[21]
beta_data = remove_outliers(beta_raw_data)
def beta_equations(p, *args):
    (a, b) = p
    (m, v) = args
    return (a/(a+b) - m, a*b/((a+b)**2 * (a+b+1)) - v)

def beta_param_estimation(samples):
    mean = samples.mean()
    var = samples.var()
    (a,b) = fsolve(beta_equations, (1,1), (mean,var), xtol=0.00001)
    return [a,b]

estimate_params(beta_data, beta_param_estimation, ["a", "b"], "Beta distribution")
```

### 13) Estimating parameters for Pareto distribution:

For column in dataset d) Connected Components integer (min:1, max:13) 

```python
pareto_data = dataset[13]
def pareto_equations(p, *args):
    (k, m) = p
    (mu, v) = args
    return (k*m/(k-1) - mu, (m**2 * k)/((k-1)**2 * (k-2)) - v)

def pareto_param_estimation(samples):
    mean = samples.mean()
    var = samples.var()
    (k,m) = fsolve(pareto_equations, (3, 0), args=(mean,var), xtol=0.001)
    return [k,m]
estimate_params(pareto_data, pareto_param_estimation, ["k", "m"], "Pareto distribution")
```

```python
mvn_data = dataset.iloc[:,[6,7,16]]
def mvn_liklihood_mean(samples):
    mean_vector = samples.mean()
    cov_matrix = samples.cov()
    liklihood = multivariate_normal.pdf(samples, mean_vector, cov_matrix)
    return liklihood.mean()
data_sizes = []
param_sets = [[]]
data_size = 100
while data_size < mvn_data.shape[0]:
    param_sets[0].append(mvn_liklihood_mean(mvn_data.iloc[:data_size, :]))
    data_sizes.append(data_size)
    data_size = data_size + 100

plot_params(data_sizes, param_sets, ["Likelihood mean"], "MVN Likelihood - Datasize")
```

```python
num_samples = 50000
alpha = (10,5,3)
dirichlet_data = pd.DataFrame(np.random.dirichlet(alpha, num_samples), columns=("X1", "X2", "X3"))

data_sizes = []
param_sets = [[]]
data_size = 100
while data_size < num_samples:
    dirichlet_likelihood = dirichlet.pdf((dirichlet_data.iloc[:data_size,:]).values.transpose(), alpha)
    param_sets[0].append(dirichlet_likelihood.mean())
    data_sizes.append(data_size)
    data_size = data_size + 100
plot_params(data_sizes, param_sets, ["Likelihood mean"], "Dirichlet Likelihood - Dataset size" )
```
